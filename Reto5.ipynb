## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def evaluarNum(num1,num2):
    """ Funcion para evaluar numeros si es mayor, menor o igual qué
        ## Parametros:
            num1, num2: enteros o flotantes
        ## Return
            regresa el numero mayor o si son iguales
    """
    if(num1==num2):
        return f"El {num1} es igual que {num2}"
    elif(num1<num2):
        return f"El {num2} es mayor que {num1}"
    else:
       return f"El {num1} es mayor que {num2}"
        
````

## Ejercicio 2:

Crear una función que reciba como parameto un texto y que devuelva en el `return` una lista que contenga cada palabra del texto.

|Entrada | Salida |
| - | - | 
| "El señor de los anillos" | ["El", "señor", "de", "los", "anillos"] |



Utiliar el metodo .split() al texto

**Ejemplo**

"Buenos días".split()

````python
def separacionPalabras(frase):
    """ Funcion para separar frases 
    ## Parametros
    palabra: String
    ## Return
    regresa la frase separa por palabras
    """

    p = frase.split()
    return p
    
````

````python
separacionPalabras("El señor de los anillos")
````

## Ejercicio 3:

Un piloto de carreras ha registrado su velocidad cada segundo en la siguiente lista:

velocidad = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]

Hacer una función que imprima en que segundo el piloto rebaso los 210 km/h


**Ejemplo de impresión**

En el segundo `9` el piloto llego a los `223` km/h




````python
def identificadorVelocidad(list):
    
    for segundo, elemento in enumerate(list):

        if elemento>=210:
            print(f"En el segundo {segundo} el piloto llego a los {elemento} km/h")
       

````

````python
velocidad = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]
identificadorVelocidad(velocidad)
````

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
def parOinpar(lista):
    for e in lista:
        if e%2 == 0:
            print("El numero es par")
        else:
             print("El numero es impar")


````

````python
numeros=[2,3,6,8,9]
parOinpar(numeros)
````
